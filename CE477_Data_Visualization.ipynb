```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import calendar
import datetime
```

```python
df = pd.read_csv('all_wizmirnet_oturum.csv', sep=',', encoding="iso-8859-9")
df.index = pd.to_datetime(df['TARIH'], format='%Y-%m-%d')
```

# **Number of Access Points per month over the years.**

```python
max_ap_locs_by_month = df.groupby(pd.to_datetime(df.index).strftime('%Y-%m'))['NumberOfAPLocations'].agg(np.max).to_dict()

plt.figure(figsize=(12, 10))

plt.title('Number of Active AP Locations')
plt.xlabel("Months")
plt.ylabel("Number of AP Locations")
plt.xticks(rotation=40, ha="right")
plt.grid(axis = 'y', color = 'gray', linestyle = '--', linewidth = 0.5)

plt.bar(max_ap_locs_by_month.keys(), max_ap_locs_by_month.values(), color='green')
plt.show()
```

```python
ap_location = 'HALKAPINAR METRO'
dfdf = df.groupby(pd.to_datetime(df.index).strftime('%Y-%m'))[ap_location].agg(np.sum)

sessions_by_month = dfdf.groupby(pd.to_datetime(dfdf.index).strftime('%B')).agg(np.mean).to_dict()

plt.figure(figsize=(12, 10))

plt.title('{}'.format(ap_location))
plt.xlabel("Months")
plt.ylabel("Number of Sessions")
plt.xticks(rotation=40, ha="right")
plt.grid(axis = 'y', color = 'gray', linestyle = '--', linewidth = 0.5)

plt.bar(sorted(sessions_by_month.keys(), key=lambda m: datetime.datetime.strptime(m, "%B")), sessions_by_month.values(), color='blue')
plt.show()
```

```python
ap_location = 'HALKAPINAR METRO'
sessions_by_date = df[ap_location].to_dict()

plt.figure(figsize=(12, 10))

plt.title('{}'.format(ap_location))
plt.xlabel("Dates")
plt.ylabel("Number of Sessions")
plt.xticks(rotation=40, ha="right")
plt.grid(axis = 'y', color = 'gray', linestyle = '--', linewidth = 0.5)

plt.plot(sessions_by_date.keys(), sessions_by_date.values(), color='blue', lw=2)
plt.show()
```
